## Peaks of $B_\nu$ and $B_\lambda$

```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns; sns.set(context="poster")
import numpy as np

plt.figure()
xs = np.linspace(0,10)
plt.plot(xs, planck_physics.d_B_nu_d_nu_dimensionless(xs))
plt.axhline(y=0, linestyle="dashed")
plt.xlabel(r"$x$")
plt.ylabel(r"$dB / dx$ [arbitrary units]")

plt.figure()
ys = np.linspace(0,10)
plt.plot(ys, planck_physics.d_B_lambda_d_lambda_dimensionless(ys))
plt.axhline(y=0, linestyle="dashed")
plt.xlabel(r"$y$")
plt.ylabel(r"$dB / dy$ [arbitrary units]")
```

```python
from IPython.display import display, Math

x_0 = 3
Wein_peak_nu_dimensionless = root_finder(planck_physics.d_B_nu_d_nu_dimensionless, x_0)
display(Math(r"B_\nu \text{{ peaks at: }} x = {0:.4} \frac{{h \nu}}{{k T}}".format(Wein_peak_nu_dimensionless)) )


y_0 = 5
Wein_peak_lambda_dimensionless = root_finder(planck_physics.d_B_lambda_d_lambda_dimensionless, y_0)
display(Math(r"B_\lambda \text{{ peaks at: }} y = {0:.4} \frac{{h c}}{{\lambda k T}}".format(Wein_peak_lambda_dimensionless)) )

```

## Peaks of $\frac{\partial B_\nu}{\partial T}$

```python
xs = np.linspace(0,10)
plt.plot(xs, planck_physics.d_B_nu_d_T_d_nu_dimensionless(xs))
plt.axhline(y=0, linestyle="dashed")
plt.xlabel(r"$x$")
plt.ylabel(r"$\frac{d^2B}{\partial x \partial T}$ [arbitrary units]")
```

```python
x_0 = 4
d_B_nu_dT_peak_dimensionless = root_finder(planck_physics.d_B_nu_d_T_d_nu_dimensionless, x_0)
display(Math(r"\frac{{\partial B_\nu}}{{\partial T}} \text{{ peaks at: }} x = {0:.4} \frac{{h \nu}}{{k T}}".format(d_B_nu_dT_peak_dimensionless)) )

```
